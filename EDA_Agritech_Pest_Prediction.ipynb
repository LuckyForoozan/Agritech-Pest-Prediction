**7 Identify Outliers**

```python
# Identifying Outliers using Z-scores

from scipy import stats

# Calculate Z-scores for each numerical column
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
outliers_z = (z_scores > 3) | (z_scores < -3)

# Display the outliers
print("Outliers using Z-scores (True represents an outlier):\n", outliers_z)

```

**8 Check Data Imbalance**

```python
#Analyze Class Distribution with value_counts()
# Check the distribution of each categorical column
for col in df.select_dtypes(include=['object']).columns:
    print(f"Class Distribution for {col}:\n", df[col].value_counts())
    print("\n")
```

```python
#Calculate Class Imbalance (Percentage Distribution)
# Calculate the percentage of each class in a categorical column
for col in df.select_dtypes(include=['object']).columns:
    class_percentage = df[col].value_counts(normalize=True) * 100
    print(f"Class Percentage for {col}:\n", class_percentage)
    print("\n")

```

**9 Explore Relationships**

```python
#Correlation Matrix (Numerical Variables)
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
numerical_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
## Pair plot
import seaborn as sns
import matplotlib.pyplot as plt

# Pair plot with only the lower triangle (removes plots beyond the main diagonal)
sns.pairplot(df, corner=True)
plt.show()

```
